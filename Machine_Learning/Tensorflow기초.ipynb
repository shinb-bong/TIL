```python
키 = [170,180,175,160]
신발 = [260,270,265,255]
# 신발 = 키 * a +b

a = tf.Variable(0.1)
b = tf.Variable(0.1)

def loss_function():
  예측 =  키 * a + b
  return tf.square(신발 - 예측)
# 경사하강법
opt = tf.keras.optimizers.Adam(learning_rate=0.1)

for i in range(300):
  # 손실함수, var_list
  opt.minimize(loss_function,var_list=[a,b])
  print(a.numpy(),b.numpy())
```

```python
train_x = [1,2,3,4,5,6,7]
train_y = [3,5,7,9,11,13,15]

# 모델 만들기 => 학습 순으로 진행
a = tf.Variable(0.1)
b = tf.Variable(0.1)

def loss_function(a,b):
  예측_y = train_x * a + b
  return tf.keras.losses.mse(train_y,예측_y)

# 경사하강법
opt = tf.keras.optimizers.Adam(learning_rate=0.01)

for i in range(1000):
  # 손실함수, var_list
  opt.minimize(lambda:loss_function(a,b),var_list=[a,b])
  print(a.numpy(),b.numpy())
```

```python
import pandas as pd

data = pd.read_csv('경로')
print(data)

print(data.isnull().sum())
data = data.dropna()
# data.fillna(100)

# 리스트형태로 담아줌
y데이터 = data['admit'].values

x데이터 = []
for i , rows in data.iterrow():
  x데이터.append([rows['gre'], rows['gpa'],rows['rank']])

import numpy as np
import tensorflow as tf

# 실험적으로 파악해서 모델을 만들어야함
# 마지막에 오는건 이때까지 공부 한 것 처럼 주의해줘야함.
model = tf.keras.models.Sequential([
                                    tf.keras.layers.Dense(64,activation='tanh'),
                                    tf.keras.layers.Dense(128,activation='tanh'),
                                    tf.keras.layers.Dense(1,activiation='sigmoid')

])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

# 학습시키기 np.array만 넣어줄수있음
model.fit(np.array(x데이터),np.array(y데이터),epochs=10)

# 예측하기
예측값 = model.predict([[750,3.70,3],[400,2.2,1]])
print(예측값)
```
